### Some plotting functions

```python
def positive_rate(data, alpha=0.05):
    return (data < alpha).sum() / len(data)

def truncate_colormap(cmap, minval=None, maxval=None, n=100):
    '''
    https://stackoverflow.com/a/18926541
    '''
    if isinstance(cmap, str):
        cmap = plt.get_cmap(cmap)
    new_cmap = mpl.colors.LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=minval, b=maxval),
        cmap(np.linspace(minval, maxval, n)))
    return new_cmap

def positive_rate_heatmap(data, col_a, col_b, value, vmin=None, vmax=None, cmapmin=0, cmapmax=0.8,
                          linecolor="gray",
                          palette="Reds", **kws):

    r2_y_yhat = data.pivot_table(index=col_b, values=['r2_y_yhat'], columns=col_a, aggfunc=np.mean)
    r2_y_c = data.pivot_table(index=col_b, values=['r2_y_c'], columns=col_a, aggfunc=np.mean)
    r2_yhat_c = data.pivot_table(index=col_b, values=['r2_yhat_c'], columns=col_a, aggfunc=np.mean)

    pr = data.pivot_table(index=col_b, values=[value], columns=col_a, aggfunc=positive_rate)

    annot = np.round(pr, 2).astype(str).values  # + '\n' + np.round(r2_y_c, 2).astype(str).values

    print(np.round(np.mean(r2_y_yhat.values), 2),
          np.round(np.mean(r2_yhat_c.values), 2),
          np.round(np.mean(r2_yhat_c.values), 2))

    sns.heatmap(data.pivot_table(index=col_b, values=[value], columns=col_a, aggfunc=positive_rate),
                linewidths=0.5,
                linecolor=linecolor,
                vmin=vmin, vmax=vmax,
                cmap=truncate_colormap(sns.color_palette(palette, as_cmap=True), cmapmin, cmapmax),
                annot=annot, #True,
                fmt='',
                #annot=True,
                **kws)
```

# Effect of conditional distribution in case of linear relationship

Weights are adjusted so that correlations y-c and y-yhat are kept around 0.3 (expl.var=10%)
and for c-yhat: 0.2 (expl.var=4%)

The tested weights are:
- e=0.1 d=2: wyc=0.1, wyyhat=0.1,  wcyhat=0.06 N=50,100,500,1000
- e=1, d=0: wyc=0.33, wyyhat=0.33, wcyhat=0.2, N=50,100,500,1000
- e=1.05, d=-3, wyc=4, wyyhat=4,    wcyhat=2.5,  N=50,100,500,1000
- e=1.5,  d=-5: wyc=55, wyyhat=55,  wcyhat=40,  N=50,100,500,1000
- e=5,   d=-10: wyc=3000000, wyyhat=3000000, wcyhat=1000000, N=50,100,500,1000

To run simulations, launch `simulated/*_simulate_normviol_*.py` form the project root.

```python
sns.set(rc={"figure.figsize":(6, 2)})
sns.set_style("whitegrid")

n=100
ci=_binom_ci(success=0.05*n, total=n, ci=.95)
print("Binomial confidence intervals for", n, "simulations (per tile):",
      np.round(ci, decimals=3))

ax = pd.read_csv('results/h0_simulate_normviol_lin.csv', index_col=0).T.plot.bar(ylim=(0,0.5), legend=False, width=.8)
ax.axhline(ci[1], color="gray", linestyle='--')
ax.grid(axis='x')
plt.savefig('../fig/raw/sim_h0_norm_violated_lin.pdf')

ax = pd.read_csv('results/h1_simulate_normviol_lin.csv', index_col=0).T.plot.bar(ylim=(0,1), legend=False, width=.8)
ax.axhline(ci[1], color="gray", linestyle='--')
ax.grid(axis='x')
plt.savefig('../fig/raw/sim_h1_norm_violated_lin.pdf')
ax = pd.read_csv('results/h0_simulate_normviol_sig.csv', index_col=0).T.plot.bar(ylim=(0,0.5), legend=False, width=.8)
ax.axhline(ci[1], color="gray", linestyle='--')
ax.grid(axis='x')
plt.savefig('../fig/raw/sim_h0_norm_violated_sig.pdf')
ax = pd.read_csv('results/h1_simulate_normviol_sig.csv', index_col=0).T.plot.bar(ylim=(0,1), legend=False, width=.8)
ax.axhline(ci[1], color="gray", linestyle='--')
ax.grid(axis='x')
plt.savefig('../fig/raw/sim_h1_norm_violated_sig.pdf')
```
